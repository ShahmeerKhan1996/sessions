### GROUP BY --- HAVING

GROUP BY works after the WHERE clause filters the rows and before the HAVING clause filters the groups.

- Count Employees by JobTitle
- Average TotalPay by JobTitle
- Employees with the Highest TotalPay in Each Year
- Sum of TotalPay by Year and JobTitle

```python
q = '''select jobtitle, count(*) as employee_count
    from Salaries
    group by jobtitle'''

sql_query(q)
```

```python
q = '''select jobtitle, avg(totalpay) as 'Average Total Pay'
    from Salaries
    group by 1'''

sql_query(q)
```

```python
q = '''select Jobtitle, sum(totalpay) as TotalPay from Salaries
    group by 1'''

sql_query(q)
```

```python
q = '''select Employeename, max(totalpay) as 'Highest Total Pay', year
    from Salaries
    group by 3'''

sql_query(q)
```

```python
q = '''select Employeename, max(totalpay) HighestSalary from Salaries
    where year == 2014'''

sql_query(q)
```

```python
q = '''select jobtitle, count(*) as employee_count
    from Salaries
    group by jobtitle order by employee_count desc'''

sql_query(q)
```

```python
q = '''select jobtitle, year, totalpay as TotalPay
    from Salaries
    group by 2'''

sql_query(q)
```

```python
q = '''select jobtitle, sum(totalpay) TotalPay from Salaries
    group by 1
    where sum(totalpay) > 1000000'''

sql_query(q)
```

- Total Pay Greater Than 1,000,000 by JobTitle
- Average BasePay Greater Than 50,000 by JobTitle
- Employee Count Greater Than 10 in Each Year
- Job Titles with More Than 5 Employees and Average TotalPay Greater Than 100,000

```python
q = '''select jobtitle, sum(totalpay) TotalPay from Salaries
    group by 1
    having sum(totalpay) > 1000000'''

sql_query(q)
```

```python
q = '''select jobtitle, avg(basepay) BasePay from Salaries
    group by 1
    having avg(basepay) > 50000'''

sql_query(q)
```

```python
q = '''select count(employeename) as EmployeeCount, Year
from Salaries
group by 2
having count(employeename) > 10'''

sql_query(q)
```

```python
q = '''select employeename, count(employeename) from salaries
    where year = 2011'''

sql_query(q)
```

### CASE

- Categorizing Employees Based on Their Total Pay
- Categorizing Employees Based on BasePay and OvertimePay
- Categorizing Job Titles Based on Pay

```python
q ='''SELECT Id, EmployeeName, BasePay, OvertimePay,
       CASE
           WHEN OvertimePay > 10000 THEN 'Overtime Worker'
           ELSE 'Regular Worker'
       END AS WorkCategory
FROM Salaries
'''

sql_query(q)
```

```python
q='''SELECT Id, EmployeeName, JobTitle, BasePay,
       CASE
           WHEN BasePay > 90000 THEN 'High Paying Job'
           WHEN BasePay BETWEEN 60000 AND 90000 THEN 'Medium Paying Job'
           ELSE 'Low Paying Job'
       END AS JobCategory
FROM Salaries'''

sql_query(q)
```

```python
q = '''select employeename, jobtitle, totalpay, basepay,
            case
                when totalpay > 50000 THEN '>50k'
                when basepay < 10000 THEN '<10K'
                else 'NaN'
            END as XXXX
        from Salaries'''

sql_query(q)
```
